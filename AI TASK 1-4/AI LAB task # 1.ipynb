```python

import random

def choose_word():
    words = ["pakistan","china","india","nepal","england","africa","bangladesh"]
    return random.choice(words)

def show_word(word, guessed_letters):
    return "".join(letter if letter in guessed_letters else "_" for letter in word)
hangman_stages = [
    """
     -----
     |   |
         |
         |
         |
         |
    =========
    """,
    """
     -----
     |   |
     O   |
         |
         |
         |
    =========
    """,
    """
     -----
     |   |
     O   |
     |   |
         |
         |
    =========
    """,
    """
     -----
     |   |
     O   |
    /|   |
         |
         |
    =========
    """,
    """
     -----
     |   |
     O   |
    /|\\  |
         |
         |
    =========
    """,
    """
     -----
     |   |
     O   |
    /|\\  |
    /    |
         |
    =========
    """,
    """
     -----
     |   |
     O   |
    /|\\  |
    / \\  |
         |
    =========
    """
]

def hangman_game():
    secret_word = choose_word()
    guessed_letters = []
    attempts = len(hangman_stages) - 1 



    print("Welcome to Hangman!")
    print("Guess the word by entering one letter at a time.")
    print("You have attempts to guess the word correctly.")
    print(show_word(secret_word, guessed_letters))
    

    while True:
        print(hangman_stages[len(hangman_stages) - 1 - attempt])

        guess_word = input("Enter your guess letter : ").lower()

        if len(guess_word) != 1 or not guess_word.isalpha():
            print("Please enter a single letter.")
            continue

        if guess_word in guessed_letters:
            print("You've already guessed that letter.")
            continue

        guessed_letters.append(guess_word)

        if guess_word not in secret_word:
            attempts -=1
            print(f"Incorrect guess! You have {attempts} attempts left.")
            hint = input("if you want hint enter y/n : ").lower()
            if hint== "y":
                print("The word is from country ")
            elif hint == "n":
                print("as your wish :)")    
            elif attempts == 0:
                print(f"Sorry, you're out of attempts! The word was '{secret_word}'.")
                break
        else:
            print("Correct guess!")

        word_progress = show_word(secret_word, guessed_letters)
        print(word_progress)

        if "_" not in word_progress:
            print("Congratulations, you've guessed the word!")
            break


hangman_game()



```
